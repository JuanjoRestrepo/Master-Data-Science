La visualización de datos es el proceso mediante el que tomamos un conjunto de datos de interés y los representamos utilizando gráficas. Su importancia radica en la forma como nuestro cerebro procesa la información. Desde siempre el ser humano se ha sentido atraído por lo visual (imágenes, colores). Por ejemplo, los primeros humanos, para comunicarse, utilizaban imágenes, las pintaban en cuevas. Las primeras formas de escritura era por medio de imágenes que tenían un significado específico en su cultura. Lo visual capta su interés y le permite centrarse en el mensaje que se quiere transmitir. Es por ello que el uso de estas gráficas (tablas, barras, tortas, histogramas, cajas y bigotes, entre otras) proporcionan una manera accesible de ver y comprender tendencias, valores atípicos y patrones en los datos. Con el auge del BigData la visualización de datos se ha convertido en una herramienta esencial.

#### Ventajas de visualizar datos
1. Herramienta clave para poder analizar grandes cantidades de datos y en formatos distintos.
2. Permite la absorción casi instantanea de gran cantidad de información, en contraposición a lo que ocurre con ver, por ejemplo, una hoja de cálculo.
3. Acelera el proces de toma de decisiones. Esto porque la información se puede comprender de mánera más rápida, permitiendo ver la relación entre operaciones y resultados.
4. Revela patrones y tendencias, vital para la inteligencia de negocios.
5. Es interactiva, permite a los usuarios manipular la información en diferentes configuraciones (con ayuda de aplicaciones).
6. Mejora la comunicación y divulgación de la información.

*Una imagen vale más que mil palabras*

#### Tipos de gráficas más utilizados:
1. **Gráficas de barras**: organizan la información en barras rectangulares facilitando la comparación de datos relacionados. Se pueden usar para comparar dos o más valores en una categoría.
2. **Graáficas de líneas**: organizan la información en un formato compacto y contínuo, facilitando la interpretación de diferentes tipos de información. Ayudan a comprender tendencias, patrones y fluctuaciones en un conjunto de datos.
3. **Diagramas de dispersión**: útiles para visualizar datos que contengan medidas cuántitativas que cambien con frecuencia.
4. **Gráficas de tortas**: fáciles de construir y entender. Usados comúnmente para mostrar la relación de las partes de un todo, lo que se conoce como porcentajes.
5. **Gráficas de bala**: utilizadas para comparar dos valores únicamente, un valor actual y sus escalas para determinar si los cambios son buenos o malos. Se usan para monitorear un valor clave.
6. **Gráficas de cajas y bigotes**: utilizadas para ver la distribución de los datos en cuartiles y detectar datos atípicos importantes llamados outliers.
7. **Mapas de calor**: usadas para representar visualmente la correlación entre datos, es decir, que tan relacionados están los datos entre sí.
8. **Gráficas de mapas**: permiten visualizar los datos distribuídos de acuerdo a su ubicación geográfica.

En este ejemplo utilizaremos el dataset de la unidad anterior **Video Game Sales with Rankings**, pero en este caso usaremos la versión que ya ha sido procesada, es decir, el resultado de la limpieza de datos que aplicamos sobre el dataset original.

Como siempre comenzamos importando los módulos necesarios para trabajar con los datos. En este caso nos centraremos más en el módulo **matplotlib**.

```python
# Importamos los módulos necesarios.
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

Importamos el dataset con ayuda de pandas. En este caso ya no vamos a hacer ningún procedimiento sobre los datos del dataset sino que lo vamos a utilizar para generar algunas gráficas.

```python
# Usamos el módulo pd para cargar el dataset y utilizamos el método head() para verificar que todo esté bien.
df = pd.read_csv("videogame_sales_with_ratings_cleaned.csv")
df.sample(5)
```

Con esto verificamos que nuestro dataset procesado carga y funciona de manera correcta. Vamos a proceder entonces con las diferentes visualizaciones que podemos realizar.

#### Visualización para análisis

Como vimos la Unidad anterior, podemos usar visualización de datos para realizar análisis exploratorio y limpieza de datos. Por ejemplo podemos visualizar la distribución de ciertos datos utilizando un histograma.

```python
# Usamos el método hist() para presentar un histograma o gráfica de barras.
plt.hist(df['Genre'])
plt.show()
```

En este caso estamos gráficando los datos de una única columna, la columna *Genre*. Este histograma nos está mostrando el número de apariciones de cada uno de los géneros en el dataset. Pero como se puede apreciar tenemos un problema de visualización de los géneros en el eje horizontal. Esto porque los nombres de los géneros se está superponiendo. Además tampoco tiene un título que nos indique qué estamos viendo. Veamos como podemos mejorarlo.

```python
# Cambiamos el estilo del gráfico.
plt.style.use('fivethirtyeight')
plt.figure(figsize=(7, 4)) # Con figure ajustamos el tamaño de la gráfica.

genre = df['Genre']

# Volvemos a usar el método hist() del módulo matplotlib.
plt.hist(genre)

# Añadimos un título a la gráfica.
plt.title('Géneros de videojuegos')

# Añadimos labels a los ejes de la gráfica.
plt.xlabel('Géneros')
plt.ylabel('Apariciones del género')

# Hacemos una rotación de los ticks (labels pequeños) del eje x.
plt.xticks(rotation='vertical')

plt.show()
```

¡Esta nueva gráfica se ve mucho mejor! Ajustamos el estilo de la gráfica y el tamaño. Añadimos un título a la gráfica para tener una idea general de los datos que estamos viendo. Adicional a esto, colocamos labels en los dos ejes que nos brindan información más detallada. Hicimos una rotación de los ticks del eje *x* de forma que no aparecieran unos sobre otros. Incluso, cambiándole el estilo, añadimos una cuadricula de fondo que ayuda a ver más claro los valores de cada dato  en  el histograma.

Ahora si queremos ver el comportamiento de una columna numérica, por ejemplo *NA_Sales*, podemos usar una gráfica de cajas y bigotes.

```python
# Obtenemos la información de las ventas en Norte América.
na_sales = df['NA_Sales']

# Usamos el método boxplot() del módulo matplotlib.
plt.boxplot(na_sales)

# Añadimos un título a la gráfica.
plt.title('Comportamiento de las ventas en N.A.')

# Añadimos los labels de cada eje de la gráfica.
plt.xlabel('Norte America')
plt.ylabel('N° de copias vendidas (MM)')

plt.show()
```

Esta gráfica nos permite ver información estadística como la mediana (dato de la mitad, representado por línea de color naranja); los cuartiles (primero y tercero) y los bigotes (inferior y superior) que, en este caso, se están correspondiendo con la mediana por lo que no son fáciles de ver; los datos atípicos que son todos aquellos que se encuentran por fuera del área entre los bigotes siendo los más alejados, los outliers. En este caso podemos decir que, como la columna analizada se refiere a ventas, los outliers se presentan porque existen videojuegos muy bien recibidos por el público que logran muchas más ventas que la mayor parte de los otros juegos. Como podemos ver, y comparando con las gráficas de cajas y bigotes vistas en el análisis exploratorio, esta se ve mucho mejor pues posee más información que permite a un usuario analista entender más fácil lo que está observando en la gráfica. 

Otra forma de visualización muy utilizada para realizar análisis son las tablas. En la Unidad anterior vimos un ejemplo usando una tabla que mostraba la correclación entre varias columnas numéricas. Veámosla nuevamente, agregando esta vez otras columnas.

```python
# Obtenemos todas las columnas de tipo numérico.
numeric = df[[
    'NA_Sales', 
    'EU_Sales', 
    'JP_Sales', 
    'Other_Sales', 
    'Global_Sales', 
    'Critic_Score', 
    'Critic_Count', 
    'User_Score', 
    'User_Count'
]]
# Usamos el método corr() para calcular y mostrar las relaciones entre las columnas.
numeric_corr = numeric.corr()
numeric_corr
```

Aunque en la tabla de correlación que vimos durante el análisis exploratorio era fácil de entender por tener cuatro columnas, en esta ocasión se torna un poco más complicado el análisis debido a la cantidad de columnas que estamos tratando de relacionar. Con el fin de ayudar un poco a la compresión de este tipo de tablas podemos usar una gráfica llamada mapa de calor.

```python
# Indicamos a mathplotlib que vamos a graficar un mapa de calor usando el método imshow().
plt.imshow(numeric_corr, cmap='coolwarm', interpolation='none')
plt.colorbar() # Indicamos que queremos mostrar la barra de color.

# Asignamos los labels que se mostraran tanto en el eje horizontal como en el vertical.
plt.xticks(range(len(numeric.columns)), numeric.columns, rotation=90)
plt.yticks(range(len(numeric.columns)), numeric.columns)

# Ajustamos el tamaño de la gráfica.
plt.gcf().set_size_inches(9, 9)

# Obtenemos todos los valores de la tabla de correlación.
labels = numeric_corr.values
for y in range(labels.shape[0]): # Recorremos cada registro (fila).
    for x in range(labels.shape[1]): # Por cada registro recorremos sus columnas para obtener los valores respectivos.
        # Con el método text() de mathplotlib dibujamos los valores para cada celda del mapa de calor.
        plt.text(x, y, '{:.2f}'.format(labels[y, x]), ha='center', va='center', color='white')

plt.show()
```

Como podemos observar este gráfico es sumamente útil para ver las relaciones entre las columnas debido a los colores. Recordemos que el coeficiente de correlación entre dos datos es un valor ubicado entre 1 y -1. Si el coeficiente de correlación es cercano a 1 entonces existe una relación lineal directa fuerte entre los dos datos, es decir, cuando el valor de uno de los datos crece, el otro también lo hace. Por el contrario si el coeficiente de correlación es cercano a -1 entonces la relación exite una relación linea inversa entre los dos datos, es decir, cuando el valor de uno de los datos crece, el otro disminuye. Si el coeficiente de correlación tiende a cero, no existe relación entre los datos. La barra de colores del lado derecho de la gráfica ayuda a interpretarla pues muestra las diferentes gamas para cada valor numérico (E.g. el color rojo corresponde al valor 1). Gracias a esto podemos, con simplemente mirar la gráfica, darnos cuenta rápidamente que hay una fuerte relación por ejemplo entre las ventas en Norte América y las ventas globales (columnas *NA_Sales* y *Global_Sales* respectivamente).

Veamos ahora como podemos usar algunas gráficas para visualizar reportes.

#### Visualización para reportes

La visualización de datos también es una herramienta sumamente útil para generar reportes que permitan la toma de decisiones. Veamos tres ejemplos de reportes que podemos visualizar mediante gráficas.
1. Para todos los géneros de videojuegos mostrar una gráfica comparativa del total de copias vendidas en Europa.
2. Para los editores Activision, Nintendo, Square Enix y Ubisoft, mostrar el porcentaje de copias vendidas con respecto al total de ventas a nivel global.
3. Para los juegos de acción, FPS y rol mostrar una gráfica comparativa del total de juegos publicados de estos géneros entre los años 2015 y 2020.

Veamos el primer caso. Aquí nos piden una comparación del total de copias vendidas para distintos géneros de videojuegos. Como es una comparación de totales una gráfica que funciona muy bien es la gráfica de barras.

```python
# Obtenemos las columnas requeridas.
sales_pg = df[['Genre', 'EU_Sales']]

# Agrupamos usando el método groupby() y sumamos los valores con el método sum().
# El parámetro as_index=False permite que podamos seguir trabajando con el dataframe completo al no convertir la columna 
# que se usa para agrupar en un índice.
eu_sales = sales_pg.groupby(['Genre'], as_index=False).sum()

# Construimos la gráfica de barras usando el método bar().
plt.bar(eu_sales['Genre'], eu_sales['EU_Sales'], color='#FF8A65')

# Asignamos el títulos a la gráfica.
plt.title('Ventas en Europa por género')

# Colocamos los labels a los ejes horizontal y vertical respectivamente.
plt.xlabel('Géneros')
plt.ylabel('N° de copias vendidas (MM)')

plt.xticks(rotation=90)

plt.show()
```

Claramente podemos apreciar que los juegos de género acción mandan la parada entre todos los géneros.

Veamos ahora el segundo caso. Se nos pide mostrar el porcentaje de copias vendidas por cada editor (Activision, Nintendo, Square Enix, Ubisoft), comparándolos con el total de copias vendidas en el mundo. Para este tipo de situaciones una buena gráfica para utilizar es la torta.

```python
# Obtenemos todos los registros correspondientes a los editores requeridos.
publisher = df['Publisher']
publisher_f = (publisher == 'Activision') | (publisher == 'Nintendo') | (publisher == 'Square Enix') | (publisher == 'Ubisoft')
p_gs = df.loc[publisher_f, ['Publisher', 'Global_Sales']]

# Obtenemos el total de copias vendidas.
total_sales = df['Global_Sales'].sum()

# Obtenemos el porcentaje de copias vendidas por editor.
publisher_names = p_gs['Publisher'].unique()
sales_percents = []
total_pp = 0
for i in range(len(publisher_names)):
    name = publisher_names[i]
    spp_percent = (p_gs[p_gs['Publisher'] == name]['Global_Sales'].sum() / total_sales) * 100
    sales_percents.append(spp_percent)
    total_pp += spp_percent

#Obtenemos el porcentaje restante de ventas.
other_percent = 100 - total_pp
publisher_names = np.append(arr=publisher_names, values=['Otros'])
sales_percents.append(other_percent)

# Usamos el método pie() para construir la gráfica de torta.
plt.figure(figsize=(8, 5))
wedges, texts, autotexts = plt.pie(sales_percents, autopct='%.0f', pctdistance=1.15)

# Asignamos un título a la gráfica.
plt.title('Porcentaje de ventas globales')

# Agregamos el recuadro de labels.
plt.legend(wedges, publisher_names, loc='right', bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```

Esta gráfica nos muestra que, con respecto al total de ventas globales:
- Nintendo goza de un 16% de las ventas globales.
- Activision y Square Enix comparten el 1% de las ventas globales.
- Ubisoft tiene el 2% de las ventas globales.

Resolvamos el último caso. Para este tenemos que mostrar una gráfica que compare el total de publicaciones de videojuegos de tres géneros específicos (acción, shooter, rol), durante los años 2015 a 2020. En este caso dónde estamos tratando de saber el comportamiento de las publicaciones a lo largo de 5 años, el gráfico más conveniente es el gráfico de líneas para hacer las comparaciones.

```python
# Convertimos los valores de la columna de años en tipos enteros.
df['Year_of_Release'] = df['Year_of_Release'].apply(np.int64)

# Obtenemos las columnas correspondientes a los géneros y años de publicación correspondientes.
releases_f = ((df['Genre'] == 'Action') | (df['Genre'] == 'Shooter') | (df['Genre'] == 'Role-Playing')) & ((df['Year_of_Release'] >= 2015) & (df['Year_of_Release'] <= 2020))
releases = df.loc[releases_f, ['Genre', 'Year_of_Release']]

# Agrupamos por género y años de publicación y usamos el método size() para contabilizar las ocurrencias.
grouped_r = releases.groupby(['Genre', 'Year_of_Release'], as_index=False).size()

# Organizamos los conjuntos que queremos graficar.
genre_names = grouped_r['Genre'].unique()
for g in range(len(genre_names)): # Recorremos cada género.
    genre = genre_names[g]    
    genre_df = grouped_r[grouped_r['Genre'] == genre] # Obtener dataframes por género.
    
    # Usamos el método plot() para construir la gráfica de líneas por cada género.
    plt.plot(genre_df['Year_of_Release'].astype(str), genre_df['size'], label=genre)

# Asignamos un título a la gráfica.
plt.title('Publicaciones por género')

# Colocamos los labels en los ejes horizontal y vertical.
plt.xlabel('Año de publicación')
plt.ylabel('N° de publicaciones')

plt.legend(bbox_to_anchor=(0.9, 0, 0.5, 1))

plt.show()
```

Lo que podemos ver rápidamente con esta gráfica es que para estos tres géneros sólo hay registros hasta el año 2017. Y decimos que solo hay registros hasta ese año porque es muy díficil que para años posteriores no se publicaran videojuegos de estos géneros. Además vemos una disminución de publicaciones considerables entre 2015 y 2017, datos que también pueden deberse a falta de registros.

Con esto hemos visto las diferentes gráficas que podemos utilizar dependiendo de la situación en la que estemos. Además pudimos ver que el módulo **matplotlib** nos da la capacidad de "embellecer" las gráficas cambiando colores y tamaños, colocando labels y leyendas que hacen que la visualización de los datos sea más amigable y fácil de comprender.
